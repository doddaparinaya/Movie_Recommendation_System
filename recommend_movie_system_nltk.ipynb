```python
import ast
```

```python
def convert(obj):
    l=[]
    for i in ast.literal_eval(obj):
        l.append(i['name'])
    return l
```

```python
data['genres']=data['genres'].apply(convert)
```

```python
data['genres']
```

```python
data['keywords']=data['keywords'].apply(convert)
```

```python
def convert3(obj):
    l=[]
    c=0
    for i in ast.literal_eval(obj):
        if c!=3:
            l.append(i['name'])
            c=c+1
        else:
            break
    return l
            
```

```python
data['cast']=data['cast'].apply(convert3)
```

```python
data['cast']
```

```python
def fetch_director(obj):
    l=[]
    for i in ast.literal_eval(obj):
        l.append(i['name'])
        break
    return l
```

```python
data['crew']=data['crew'].apply(fetch_director)
```

```python
data['crew']
```

```python
data['overview'][0].split()
```

```python
data['overview']=data['overview'].apply(lambda x:str(x).split())
```

```python
data['genres']=data['genres'].apply(lambda x:[i.replace(' ','')for i in x])
data['keywords']=data['keywords'].apply(lambda x:[i.replace(' ','')for i in x])
data['cast']=data['cast'].apply(lambda x:[i.replace(' ','')for i in x])
data['crew']=data['crew'].apply(lambda x:[i.replace(' ','')for i in x])
```

```python
data.columns
```

```python
data['tags']=data['overview']+data['genres']+data['keywords']+data['crew']+data['cast']
```

```python
new_data=data[['movie_id','title','tags']]
```

```python
new_data.head()
```

```python
new_data['tags']=new_data['tags'].apply(lambda x:' '.join(x))
```

```python
new_data['tags'][0]
```

```python
new_data['tags']=new_data['tags'].apply(lambda x:x.lower())
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector=cv.fit_transform(new_data['tags']).toarray()
```

```python
vector[0]
```

```python
cv.get_feature_names()
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return ' '.join(y)
```

```python
new_data['tags']=new_data['tags'].apply(stem)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vector)
```

```python
similarity
```

```python
similarity[0]
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    movie_index=new_data[new_data['title']==movie].index[0]
    distances=similarity[movie_index]
    movie_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    for i in movie_list:
        print(new_data.iloc[i[0]].title)
```

```python
recommend('Avatar')
```
